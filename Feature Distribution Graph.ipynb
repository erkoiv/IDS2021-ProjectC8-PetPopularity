```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Lodaing the data
import pandas as pd
df = pd.read_csv("./FeatureDistribution.csv")
df
```

```python
labels = df['Features']
val_80_100 = df['80-100']
val_50_80 = df['50-80']
val_30_50 = df['30-50']
val_0_30 = df['0-30']

width = 0.60 

fig, ax = plt.subplots(figsize = (15,10))

ax.bar(labels, val_80_100, width, label='80-100')
ax.bar(labels, val_50_80, width, bottom =val_80_100,  label='50-80' )
ax.bar(labels, val_30_50, width,bottom = (np.array(val_80_100)+np.array(val_50_80)),
       label='30-50')
ax.bar(labels, val_0_30, width,bottom = (np.array(val_80_100)+np.array(val_50_80) + np.array(val_30_50)),
       label='0-30')

ax.set_ylabel('PawPularity')
ax.set_title('PawPularity Distribution By Features')
ax.legend()

plt.show()
fig.savefig('featureDistribution.png')
plt.close(fig)
```
